```python
import csv
import numpy as np
import pandas as pd
import re
```

```python
part1 = pd.read_csv('car_listings.csv')
part2 = pd.read_csv('car_listings2.csv')
part3 = pd.read_csv('car_listings3.csv')
part4 = pd.read_csv('car_listings4.csv')
part1.shape
part2.shape
part3.shape
part4.shape
```

```python
full_df = pd.concat([part1,part2,part3,part4],ignore_index=True)
full_df.shape
```

```python
full_df['year']= [full_df['model_year'][i][:4] for i in range(len(full_df))]
```

```python
full_df.year = [int(full_df.year[j]) for j in range(len(full_df))]
```

```python
full_df['model']= [full_df['model_year'][i][4:] for i in range(len(full_df))]
full_df.sample(5)
```

```python
full_df.price = [int(re.sub(',','',full_df.price[i])) for i in range(len(full_df))]
type(full_df.price[100])
full_df.sample(5)
```

```python
pd.__version__
```

```python
full_df.drop(columns='model_year', inplace=True)
```

```python
full_df.dropna(inplace=True)
full_df.shape
```

```python
full_df.isnull().sum()
```

```python
new_column_order = ['model','year','price','color','mileage','body_type','engine','description','damage','damage_det','owners',
                   'owners_det','usage','usage_det','service','service_det','dealership']
final_df = full_df[new_column_order]
final_df.sample(10)
```
